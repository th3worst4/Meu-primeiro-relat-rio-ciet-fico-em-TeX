```python
import matplotlib.pyplot as plt
import numpy as np

hfont = {'fontname':'Times New Roman'}

t  = np.loadtxt("./Dados_diagrama _inicial.txt",usecols=0) 
y  = np.loadtxt("./Dados_diagrama _inicial.txt",usecols=1)

m_t = [0.08373333333332994, 0.3182000000000007 ,0.5191555555555569]
m_y = [0.3515142905963731, 0.03180401117815834 ,0.2001234361148974]
```

```python
figura = plt.figure(figsize=(7,5.5)) #tamanho da figura
ax = figura.add_subplot(111)

ax.set_title('Gráfico Deslocamento x Tempo',fontsize=15, **hfont)

ax.grid(visible=True,which="both",axis="both")
ax.set_ylabel("Altura(m)",fontsize=15, **hfont)
ax.set_xlabel("Tempo(s)",fontsize=15, **hfont)

plt.box(on=True)
ax.plot(t,y)
ax.plot(m_t,m_y, 'o')
#ax.errorbar(max_t,max_y, yerr = 0.0005, xerr = None, fmt =  'o', capsize = 5,  label="pontos de altura máxima")

plt.text(0.08373333333332994+0.02, 0.3182000000000007+0.03, r'$h_0 \Rightarrow K=0 J$', fontsize=15,**hfont)
plt.text(0.3515142905963731-0.17, 0.03180401117815834-0.002, r'$h=0 \Rightarrow U_g =0 J$', fontsize=15,**hfont)
plt.text(0.5191555555555569-0.02, 0.2001234361148974+0.005, r'$h_1$', fontsize=15,**hfont)
plt.axhline(0.03180401117815834, 0, 15,color="darkorange",linestyle="-.")

plt.savefig('diagrama_inicial.png', dpi=300, bbox_inches='tight') #salvar


plt.show()
```
